# FRECUENCIA POR AÑOS DE 5 PAISES DESPUES DEL TERCER PAIS CON MAS FRECUENCIA POR AÑO (Grafica de cada país)

```python
# Seleccionar los años que deseas incluir en el gráfico
años_seleccionados = [2010, 2011, 2012, 2013, 2014]

# Filtrar el DataFrame para incluir solo los años seleccionados
df_filtrado = ufo_df_new[ufo_df_new["Year"].isin(años_seleccionados)]

# Contar la frecuencia de países en cada año
conteo_paises = df_filtrado.groupby(["Year", "Country"]).size().unstack().fillna(0)

# Seleccionar solo los países después del tercero solo los 5 siguientes
paises_despues_del_tercero = conteo_paises.iloc[:, 3:8]

# Utilizar el estilo de Seaborn
sns.set(style="whitegrid")

# Crear subgráficos para cada país
fig, axes = plt.subplots(nrows=len(paises_despues_del_tercero.columns), ncols=1, figsize=(10, 5 * len(paises_despues_del_tercero.columns)))

# Iterar sobre cada país y crear un gráfico por país
for i, country in enumerate(paises_despues_del_tercero.columns):
    ax = axes[i]
    paises_despues_del_tercero[country].plot(kind="bar", stacked=True, ax=ax)
    ax.set_xlabel('Año')
    ax.set_ylabel('Frecuencia')
    ax.set_title(f'País: {country}')
    ax.legend(title='Año', loc='upper right', bbox_to_anchor=(1.2, 1))

# Ajustar la disposición de los subgráficos
plt.tight_layout()
plt.show()
```

# Es una representación visual de la frecuencia de avistamientos de OVNIs por año y mes

```python
'''
El gráfico de calor que te he proporcionado es una representación visual de la frecuencia de avistamientos de OVNIs por año y mes, limitado a los años 2011 al 2014.

Eje X (horizontal): Representa los meses del año, etiquetados del 1 al 12.
Eje Y (vertical): Representa los años, etiquetados como 2011, 2012, 2013 y 2014.
Color de las Celdas: Cada celda en el gráfico representa la frecuencia de avistamientos para un año y mes específicos. Cuanto más oscuro sea el color de la celda, mayor será la frecuencia de avistamientos para esa combinación de año y mes.
Número Anotado en las Celdas: Dentro de cada celda, hay un número que indica la cantidad de avistamientos de OVNIs registrados para ese año y mes.
El objetivo de este tipo de visualización es proporcionar una visión general de cómo varía la frecuencia de avistamientos a lo largo de los meses y años. Puedes identificar patrones estacionales o tendencias a lo largo de los años observando cómo cambian los colores a lo largo del tiempo. Las celdas más oscuras indican períodos con mayor actividad de avistamientos, mientras que las más claras indican períodos con menos actividad.

En resumen, este gráfico de calor te permite identificar visualmente cuándo y en qué momentos del año hubo más reportes de avistamientos de OVNIs, proporcionando una representación intuitiva de la distribución de estas observaciones a lo largo del tiempo.

'''

filtered_data = ufo_df_new[(ufo_df_new["Year"] >= 2011) & (ufo_df_new["Year"] <= 2014)]

# Crear un group by para contar la frecuencia de observaciones por año y mes
grouped_data = filtered_data.groupby(["Year", "Month"]).size().unstack().fillna(0)

# Crear un gráfico de calor
sns.heatmap(grouped_data, cmap="YlGnBu", annot=True, fmt="g", linewidths=.5)
plt.title('Gráfico de Calor: Frecuencia de Avistamientos de OVNIs (2011-2014) por Año y Mes')
plt.xlabel('Mes')
plt.ylabel('Año')
plt.show()
```
